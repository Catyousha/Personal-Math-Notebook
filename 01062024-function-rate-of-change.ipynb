# $\verb|Function's Rate of Change|$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def init_graph():
    ax = plt.gca()
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_position('zero')
    ax.spines['left'].set_position('zero')
    ax.spines['right'].set_color('none')

    ax.grid(which='both', color='grey', linewidth=1, linestyle='-', alpha=0.2)

    x_ticks = np.arange(-100, 100,  1)
    y_ticks = np.arange(-100, 100, 1)
    ax.set_xticks(x_ticks[x_ticks != 0])
    ax.set_yticks(y_ticks[y_ticks != 0])
```

## $\verb|Slope of a Function|$

$$\begin{align}
    y = mx + c
\end{align}$$
$
m = \verb|slope|\\
c = \verb|offset from |(x, 0)
$

```python
def modify_slope(x):
    _m = 2
    return _m * x

def modify_offset(x):
    _b = 2
    return x + _b
 
init_graph()
x = np.linspace(-5, 5) # [-5, -4, ..., 4, 5]

y = modify_slope(x)
plt.plot(x, y, 'b', linewidth=2)

yb = modify_offset(x)
plt.plot(x, yb, 'r', linewidth=2)

# tangent point (same y):
# 2x = x+2
# 2x - x = 2
# x = 2
plt.plot(2, modify_slope(2), 'ro');
plt.annotate('(2,4)', (2,modify_offset(2)))
```

$
y = 2x \verb| (Blue) is 'sloper'|\\
y = x + 2 \verb| (Red) is higher|\\
$

## $\verb|Parabola Graph|$

$
ax^2 - bx + c\\
$

$
\verb|Example:|\\
\frac{1}{2}x^2 - 2x + 1
$

```python
init_graph()
x = np.linspace(-3, 5)

def f(_x):
    return 0.5 * _x**2 - 2*_x + 1

# 1/2 x^2 - 2x + 1
plt.plot(x, f(x), 'r',)

for i in range(-3, 6):
    y = f(i)
    plt.plot(i, y, 'bo')
    plt.annotate(f'[{i}, {y}]', (i, y))
```

Differential of parabola equation is its slope.

$m$ of $\frac{1}{2}x^2 - 2x + 1 = x - 2$

```python
init_graph()
plt.plot(x, x-2, 'r',)
```

Parabola is bouncing when x = 2.

Meanwhile in $x - 2$, $x = 2$ will produce $y = 0$.

$+y = +m$

$-y = -m$
